**This notebook show how the check the __doc__, np.source(), and github checklist of a function/method**

1. `gradient`: The gradient is computed using second order accurate central differences in the interior and either first differences or second order accurate one-sides (forward or backwards) differences at the boundaries. The returned gradient hence has the same shape as the input array.
    For two dimensional arrays, the return will be two arrays ordered by axis. In this example the first array stands for the gradient in rows and the second one in columns direction:
    
Find the formula from the github changelist.
   
* edge_order = 2:

    ```
      The one-sided boundary gradient uses 3 elements each with a different weight. A
      forward difference is used for the first element,
         dy/dx = f'(0) ~ (3.0*y[0] - 4.0*y[1] + y[2]) / (2.0*dx)
      and backwards difference is used for the last element,
         dy/dx = f'(-1) ~ (3.0*y[-1] - 4.0*y[-2] + y[-3]) / (2.0*dx)
    ```
    
* edge_order = 1:

    ```
      The one-sided boundary gradient uses 2 elements each with a different weight. A
      forward difference is used for the first element,
         dy/dx = f'(0) ~ (y[1] - y[0]) / dx
      and backwards difference is used for the last element,
         dy/dx = f'(-1) ~ (y[-1] - y[-2]) / dx
    ```

````python
from numpy import *
x = np.array([-4, -3, -2, -1, 0, 1, 2, 3, 4])
dx = np.gradient(x)
print dx

y = x**2
dy0 = gradient(y, edge_order=1)
dy = gradient(y, dx, edge_order=2)
print dy0, dy
dSgn = sign(dy)
print dSgn
d_dSgn = gradient(dSgn)
print d_dSgn

# that's the line, you need:
a = diff(sign(diff(y))).nonzero()[0] + 1 # local min+max
b = (diff(sign(diff(y))) > 0).nonzero()[0] + 1 # local min
c = (diff(sign(diff(y))) < 0).nonzero()[0] + 1 # local max
print b
````
